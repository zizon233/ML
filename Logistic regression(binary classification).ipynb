```python
import tensorflow as tf
import numpy as np
import collections
```

```python
data = np.loadtxt('Data/HMEQ.csv', delimiter = ',', dtype = str)
```

```python
y_data = data[1:,[0]]
x_data = data[1:,1:]
```

```python
li = np.array([])
```

```python
for i in range(len(x_data)):
    if '' in np.unique(x_data[i]) or '' in np.unique(y_data[i]):
        li = np.append(li,i)
```

```python
x_data = np.delete(x_data, li, 0)
y_data = np.delete(y_data, li, 0)
```

```python
for i in range(3,5):
    #print(x_data[:,i:i+1])
    li = np.unique(x_data[:,i:i+1])
    dic = {}
    index = 1
    for j in li:
        dic[j] = index
        index += 1
    for j in li:
        x_data[: , i:i+1] = np.where(x_data[:,i:i+1] == j, dic[j] , x_data[:,i:i+1])
```

```python
split = int(7*(len(x_data))/10)
#split = len(x_data)
y_train = y_data[:split]
y_test = y_data[split:]
x_train = x_data[:split]
x_test = x_data[split:]
```

```python
size = len(x_data[0])
lr = 0.00000001
```

```python
X = tf.placeholder(tf.float32, [None,size], name = 'X')
Y = tf.placeholder(tf.int32, [None,1], name = 'Y')
Y_onehot = tf.one_hot(Y,2)
print(Y_onehot)
Y_onehot = tf.reshape(Y_onehot, [-1,2])
print(Y_onehot)
W = tf.get_variable('W', [2,size])
b = tf.get_variable('b', initializer = tf.random_normal([1]))
```

```python
h = tf.matmul(X,W, transpose_b = True) + b
```

```python
hypothesis = tf.nn.softmax(h)
```

```python
#cost = -tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*tf.log(1-hypothesis))
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits = h, labels=Y_onehot))
train = tf.train.GradientDescentOptimizer(lr).minimize(cost)
```

```python
prediction = tf.argmax(hypothesis, 1)
correct_prediction = tf.equal(prediction, tf.argmax(Y_onehot, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for i in range(50001):
    loss, _, ACCURACY = sess.run([cost, train, accuracy], feed_dict = {X:x_train, Y:y_train})
    #sess.run(train, feed_dict={X:x_train, Y:y_train})
    if i%100 == 0:
        print(i, 'loss :',loss, ACCURACY)
```

```python
pred = sess.run(prediction, feed_dict = {X:x_test})
for p,y in zip(pred, y_test.flatten()):
    print(int(y)==p, 'Prediction :', p, 'True :', int(y))
```

```python
matrix = tf.confusion_matrix(y_test.flatten().astype('int64'), pred)
```

```python
matrix = matrix.eval(session=sess)
print(matrix)
```

```python
correct = matrix[0][0] + matrix[1][1]
sums = sum(sum(matrix))
```
